```python
import json
from pytubefix import YouTube
from pytubefix.cli import on_progress

dataset = "train"
with open(f"/root/project/videoqa/inputs/MedVidQA/{dataset}.json", "r") as rfile:
    data_items = json.load(rfile)
```

```python
video_url = [elem["video_url"] for elem in data_items]
video_url = list(set(video_url))
vid_num = len(video_url)

for i, vid in enumerate(video_url):
    print(f"Vid: {i}/{vid_num}")
    vid_id = vid.split("watch?v=")[-1]

    yt = YouTube(vid, on_progress_callback=on_progress)
    try:
        ys = yt.streams.get_highest_resolution()
        ys.download(output_path=f"{dataset}_set", filename=f"{vid_id}.mp4")
    except:
        print(vid)
        continue
```
